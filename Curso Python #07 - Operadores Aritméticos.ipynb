Exercicio Antecessor&Sucessor

```python
N = int(input('Digite um Número'))
S = N + 1
A = N - 1
print('Seu Número digitado foi: {}!'.format(N))
print('O Sucessor do seu número é: {}!' .format(S))
print('O Antecessor do seu número é: {}!' .format(A))
```

Exercicio que mostra o Dobro,Triplo de um valor

```python
import math

N = int(input('Digite um Número\n'))

D = N * 2
T = N * 3
R = round(math.sqrt(N),2)

print('O dobro do seu valor é: {}!'.format(D))
print('O Triplo do seu valor é: {}!'.format(T))
print('A Raiz Quadrada do seu valor é: {}!'.format(R))
```

Exercicio que lê 2 Notas e mostra a média

```python
import numpy

N1 = float(input('Digite sua 1° nota:'))
N2 = float(input('Digite sua 2° nota:'))
S = N1 + N2
M = (N1 + N2) / 2
print('A soma de suas 2 notas é: {}!'.format(S) )
print('A média de suas notas são: {}!'.format(M))
```

Exercicio Reajuste Salarial

```python
salario = float(input('Informe seu salário: R$'))
aumento = salario * 15 / 100
novo_salario = salario + aumento
print('Seu novo salario com aumento de 15% agora passa a ser R$ {:.2f}.'.format(novo_salario))
```

Exercicio de aluguel de carros

```python
quilometros = float(input('Quantos quilometros você rodou com o carro:'))
dias_alugado = int(input('Quantos dias ele ficou alugado:'))
total_dia = dias_alugado * 60
total_quilimetros = quilometros * 0.15
resposta = total_dia + total_quilimetros
print('Suas dispesas com o aluguel do carro ficou em R${:.2f}'.format(resposta))
```
